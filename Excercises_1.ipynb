[Свертка списка](https://ru.wikipedia.org/wiki/%D0%A1%D0%B2%D1%91%D1%80%D1%82%D0%BA%D0%B0_%D1%81%D0%BF%D0%B8%D1%81%D0%BA%D0%B0) -  это обобщенная операция над списком, c помощью которой можно преобразовать список в единое значение. Например рассмотрим реализации свертки слева и свертки справа (левоассоциативную свертку и правоассоциативную свертку):

```python
def foldl(f, x0, lst):
    if not lst:
        return x0
    return foldl(f, f(x0, lst[0]), lst[1:])

def foldr(f, x0, lst):
    if not lst:
        return x0
    return f(lst[0], foldr(f, x0, lst[1:]))
```

```python
#сумма
foldl(lambda x, y: x + y, 0, [1, 2, 3])
```

```python
#конкатенация
l = foldl(lambda x, y: '{}{}'.format(x, y), '', [1, 2, 3])
r = foldr(lambda x, y: '{}{}'.format(x, y), '', [1, 2, 3])
print(l, r)
```

нужно написать функцию, которая принимает две строки и проверяет, входит ли хотя бы одна перестановка второй строки в первую. Например:

> a = 'abcrotm'
> 
> b = 'tro'

функция def check_inv(a, b) вернет True, так как 'rot' содержится в 'abcrotm'. Нужно подумать как можно более оптимальный алгоритм и оценить его сложность. 

```python
a = 'abcrotmagdag'
b = 'agmd'

def check_inv(string, pattern):
    n = len(string)
    m = len(pattern)
    
    if m > n:
        return False
    
    h_s = sum([hash(s) for s in pattern])
    h_c = sum([hash(s) for s in string[0:m]])
    
    for i in range(m,n+1):
            
        if h_c == h_s:
            return (True, i)
        
        try:
            h_c = h_c - hash(string[i-m]) + hash(string[i])
        except KeyError:
            return False
        

print(check_inv(a,b))
```

# Задание 3

Реализовать бинарное дерево (класс Tree), в нём методы __repr__, __str__, __iter__ (итерация только по листьям).

```python
class Tree:
    def __init__(self, value=None, left=None, right=None):
        self.left = left
        self.right = right
        self.value = value
            
    def function(self):
        if self.left is not None:
            yield from self.left.function()
        if self.right is not None:
            yield from self.right.function()
        if (self.left is None) and (self.right is None):
            yield self.value
            
    def __iter__(self):
        if self.left is not None:
            yield from self.left.__iter__()
        if self.right is not None:
            yield from self.right.__iter__()
        if (self.left is None) and (self.right is None):
            yield self.value
        
    def __str__(self):
            
        if (self.left is None) and (self.right is None):
            return ' ' + str(self.value)
        else:
            return (str(self.left) + str(self.right)).strip()
    
    def __repr__(self, level=0):
        if not (self.left is None):
            left = 'ln: ' + self.left.__repr__(level + 1) + ';'
        else:
            left =  ''
        if not (self.right is None):
            right = 'rn: ' + self.right.__repr__(level + 1) + ';'
        else:
            right = ''
            
        c = 'value: {}; on tree level: {}'.format(self.value, level)
        
        return '''{}; {} {}'''.format(c, left, right).strip()
            
    
tree = Tree(0, Tree(1, Tree(3), Tree(4)),                             
               Tree(2))

list(tree) == [3, 4, 2]
```

# Задание 4

Реализовать простейший калькулятор математических выражений:
- только целые числа
- **+**, **\-**, **\***, **\**
- скобки

> алгоритм сортировочной станции

> pcre

```python
def calc(expr):
    op_queue = []
    rpn_queue = []
    # sorting station
    
    for sym in expr.split(' '):
        if sym in ['+', '-', '/', '*']:
            if len(op_queue):
                op = op_queue.pop()

                while op:
                    if (sym in ['*','/']) and (op in ['*','/']):
                        rpn_queue.append(op)
                    elif (sym in ['+','-']) and (op in ['+', '-']):
                        rpn_queue.append(op)
                    elif (sym in ['+', '-']) and (op in ['*', '/']):
                        rpn_queue.append(op)
                    else:
                        op_queue.append(op)
                        op_queue.append(sym)
                        break
                    
                    try:
                        op = op_queue.pop()
                    except IndexError:
                        op_queue.append(sym)
                        break
            else: 
                op_queue.append(sym)

        elif sym == '(':
            op_queue.append(sym)
        elif sym == ')':
            op = op_queue.pop()
            while op != '(':
                rpn_queue.append(op)
                op = op_queue.pop()            
        else:
            rpn_queue.append(int(sym))
            
    for op in op_queue[::-1]:
        rpn_queue.append(op)

    # calculating rpn
    stack = []    
        
    for t in rpn_queue:
        if t in ['*', '+', '/', '-']:
            try:
                [b, a] = [stack.pop(), stack.pop()]
            except (KeyError, IndexError):
                raise Exception("invalid expression")
            if   t == '*': stack.append(a * b)
            elif t == '/': stack.append(a / b)
            elif t == '+': stack.append(a + b)
            else :         stack.append(a - b)
        else:
            stack.append(t)
    
    if len(stack) > 1:
            raise Exception("invalid expression")
    return stack[0]


print(calc('2 + 3 * ( 5 - 6 ) / 3'))
print(calc('2 * ( 15 - 3 * 4 ) - 2') == 4)
print(calc( '2 / 3 * 4' ))
```
